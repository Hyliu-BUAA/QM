```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

```python
num_grids = 201
xs_lst = np.linspace(-5, 5, num_grids)
ys_lst = np.sin(xs_lst)
```

```python
dx = np.diff(xs_lst).mean()

dia = -2 * np.ones(num_grids)
offdia = np.ones(num_grids - 1)
d2grid = np.mat( np.diag(dia, 0) + np.diag(offdia, -1) + np.diag(offdia, 1) ) / np.power(dx, 2)

derivative_x_2 = -1 / 2 * d2grid
derivative_x_2
```

```python
eig_vals_lst, eig_vecs_none_lst = np.linalg.eigh(derivative_x_2)

for idx in range(5):
    plt.plot(xs_lst, eig_vecs_none_lst[:, idx],
            label="idx={0}".format(idx))

plt.title("%d Eigenfunctions"%(5))
plt.legend()
```
